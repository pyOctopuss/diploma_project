```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime, timedelta

import os
from glob import glob
```

```python
companies = ["AMAZON", "APPLE", "META", "GOOGLE", "NETFLIX"]
frequency = "daily"
path = os.getcwd()

files = glob(f"{path}/*_{frequency}.csv")
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m-%d')
```

```python
columns = ['date', 'open_price', 'highest_price', 'lowest_price', 'close_price', 'adj_close_price', 'volume']

meta = pd.read_csv(
    files[2], 
    parse_dates=['Date'],
    date_parser=dateparse
)
meta.columns = columns
meta = meta.set_index('date')
```

```python
meta['year'] = meta.index.year
meta['month'] = meta.index.month
meta['day'] = meta.index.day
meta['day_of_week'] = meta.index.weekday
meta['week_of_year'] = (meta.index.isocalendar()['week']).astype('int')
```

```python
# sns.heatmap(meta.corr(), annot=True)
```

```python
meta
```

```python
spotify_path = glob(f"{path}/spotify-2023.csv")[0]
spotify = pd.read_csv(spotify_path, encoding='latin-1')
```

```python
dates = []
for row in range(spotify.shape[0]):
    y, m, d = spotify.iloc[row]['released_year'], spotify.iloc[row]['released_month'], spotify.iloc[row]['released_day']
    dates.append(datetime(y, m, d))
    
spotify['release_date'] = dates
```

```python
# sns.heatmap(spotify[['artist_count', 'in_spotify_playlists',
# 'in_spotify_charts', 'streams', 'bpm',
# 'key', 'mode', 'danceability_%', 'valence_%', 'energy_%',
# 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']].corr(), annot=True)

spotify = spotify.sort_values('release_date', ascending=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

import xgboost
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor(**{
    'n_estimators': 500,
    'n_jobs': -1,
    'max_depth': 6,
    'eta': 0.3,
    'booster': 'gbtree',
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'subsample': 1,
    'colsample_bytree': 1,
    'min_child_weight': 1,
    'random_state': 7,
})
```

```python
# X_train = meta.loc[:'2023-10-01', list(set(meta.columns)-set(['close_price', 'adj_close_price']))]
X_train = meta.loc[:'2023-10-01', ['open_price', 'year', 'month', 'day', 'day_of_week', 'week_of_year']]
y_train = meta.loc[:'2023-10-01', ['close_price']]

# X_test = meta.loc['2023-10-01':, list(set(meta.columns)-set(['close_price', 'adj_close_price']))]
X_test = meta.loc['2023-10-01':, ['open_price', 'year', 'month', 'day', 'day_of_week', 'week_of_year']]
y_test = meta.loc['2023-10-01':, ['close_price']]

y_pred = model_xgb.fit(X_train, y_train).predict(X_test)
```

```python
y_pred_df = pd.DataFrame(y_pred, index=y_test.index)

plt.plot(meta.loc['2023-10':, ['close_price']])
plt.plot(y_pred_df)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
err = pd.DataFrame({
    'abs_err': abs(y_pred_df[0] - y_test['close_price']),
    'rel_err': abs(y_pred_df[0] / y_test['close_price'] - 1) * 100
})
err
```

```python
# X_train = meta.loc[:'2023-10-01', list(set(meta.columns)-set(['close_price', 'adj_close_price']))]
X_train = meta.loc[:'2023-10-01', ['open_price', 'highest_price', 'lowest_price', 'close_price', 'year', 'month', 'day', 'day_of_week', 'week_of_year']]
y_train = meta.loc[:'2023-10-01', ['volume']]

# X_test = meta.loc['2023-10-01':, list(set(meta.columns)-set(['close_price', 'adj_close_price']))]
X_test = meta.loc['2023-10-01':, ['open_price', 'highest_price', 'lowest_price', 'close_price', 'year', 'month', 'day', 'day_of_week', 'week_of_year']]
y_test = meta.loc['2023-10-01':, ['volume']]

y_pred = model_xgb.fit(X_train, y_train).predict(X_test)
```

```python
y_pred_df = pd.DataFrame(y_pred, index=y_test.index)

plt.plot(y_test, label='fact')
plt.plot(y_pred_df, label='pred')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

```python
err = pd.DataFrame({
    'abs_err': abs(y_pred_df[0] - y_test['volume']),
    'rel_err': abs(y_pred_df[0] / y_test['volume'] - 1) * 100
})
err
```

```python
model_xgb = XGBRegressor(**{
    'n_estimators': 500,
    'n_jobs': -1,
    'max_depth': 6,
    'eta': 0.3,
    'booster': 'gbtree',
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'subsample': 1,
    'colsample_bytree': 1,
    'min_child_weight': 1,
    'random_state': 7,
})

X_train_sp = spotify[spotify['release_date'] < datetime(2023, 7, 1)][['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']]
y_train_sp = spotify[spotify['release_date'] < datetime(2023, 7, 1)]['in_spotify_charts']

X_test_sp = spotify[spotify['release_date'] >= datetime(2023, 7, 1)][['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']]
y_test_sp = spotify[spotify['release_date'] >= datetime(2023, 7, 1)]['in_spotify_charts']

y_pred_sp = model_xgb.fit(X_train_sp, y_train_sp).predict(X_test_sp)

y_pred_sp_df = pd.DataFrame(y_pred_sp, index=y_test_sp.index)

plt.scatter(spotify[spotify['release_date'] >= datetime(2023, 7, 1)]['track_name'], y_test_sp)
plt.scatter(spotify[spotify['release_date'] >= datetime(2023, 7, 1)]['track_name'], y_pred_sp_df)
```
